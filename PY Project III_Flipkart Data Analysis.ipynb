```python
#calculate average ratings
avg_ratings=df['Ratings'].mean()
print(f'average rating:{avg_ratings}')
```

```python
#plot 1
#distribution of ratings
plt.figure(figsize=(8,6))
sns.histplot(df['Ratings'],kde=True,bins=10)
plt.title('Distribution of Ratings')
plt.xlabel('Ratings')
plt.ylabel('Frequency')
plt.savefig('flipkart1')
plt.show()
```

```python
#plot 2
#bar plot of average discount by brand
plt.figure(figsize=(10,9))
sns.barplot(x='Brand',y='Discount',data=df,estimator=np.mean)
plt.title('average discount by brand')
plt.xlabel('Brand')
plt.ylabel('Average Discount(%)')
plt.xticks(rotation=45)
plt.savefig('flipkart2')
plt.show()
```

```python
#plot 3
#count plot of number of products by brand
plt.figure(figsize=(10,6))
sns.barplot(x='Brand',data=df)
plt.title('Number of Products by Brand')
plt.xlabel('Brand')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
#plot 4
#average rating for top 5 brands by Count
top_brands=df['Brand'].value_counts().head().index
plt.figure(figsize=(10,6))
sns.barplot(x='Brand',y='Ratings',data=df[df['Brand'].isin(top_brands)],estimator=np.mean)
plt.title('Average Rating for Top 5 Brands')
plt.xlabel('Brand')
plt.ylabel('Average Rating')
plt.xticks(rotation=45)
plt.show()
```

```python
#plot 5
#Convert the MSP column to numeric (this will convert invalid parsing to NaN)
df['MSP']=pd.to_numeric(df['MSP'],errors='coerce')

#Now drop rows with NaN values in MSP column if needed
df=df.dropna(subset=['MSP'])

#Define price range segments
bins=[0,10000,20000,30000,50000,100000]#define ranges
labels=['low-end(<10k)','Budget(10k-20k)','Mid-range(20k-30k)','premium(30k,50k)','flagship(>50k)']

#create a new column for price range
df['Price_Range']=pd.cut(df['MSP'],bins=bins,labels=labels)

#count of mobiles in each price range
price_range_count=df['Price_Range'].value_counts()
plt.figure(figsize=(8,6))
sns.barplot(x=price_range_count.index,y=price_range_count.values)
plt.title('Mobile Count in Different Price Segments in India')
plt.xlabel('Price Range')
plt.ylabel('Number of Mobiles')
#adjust layout to avoid overlap
plt.tight_layout()
plt.savefig('flipkart3')
plt.show()
```

```python
#plot 06
#group by brand and get the maximum discount for each brand
brand_hightest_discount=df.groupby('Brand')['Discount'].max().reset_index()
top_brand_by_discoun=brand_hightest_discount.sort_values(by='Discount',ascending=False).head(5)
plt.figure(figsize=(10,6))
sns.barplot(x='Brand',y='Discount',data=top_brand_by_discoun,palette='coolwarm',hue='Brand',legend=False)
plt.title('Mobile Count in Different Price Segments in India')
plt.xlabel('Brand')
plt.ylabel('Discount amount')
plt.savefig('flipkart4')
plt.tight_layout()
plt.show()
```

```python
#plot 7
#top 5 brands by Average ratings
df.groupby('Brand')['Ratings'].mean().nlargest(5).plot(kind='bar')
plt.title('top 5 brands by average ratings')
plt.ylabel('average ratings')
plt.xticks(rotation=45,ha='right')
plt.tight_layout()
plt.show()
```

```python
#plot 8
#calculate the top 5 brands by total number of reviewe
top_5_brands_by_reviews=df.groupby('Brand')['No_of_reviews'].sum().sort_values(ascending=False).head()
top_5_brands_by_reviews_df=top_5_brands_by_reviews.reset_index()
#plot the top 5 brands by total number of reviews
sns.barplot(x='Brand',y='No_of_reviews',data=top_5_brands_by_reviews_df,palette='viridis',hue='Brand',legend=False)
plt.title('top 5 brands by total number of reviews')
plt.xlabel('brand')
plt.ylabel('total number of reviews')
plt.savefig('flipkart5')
plt.tight_layout()
plt.show()
```
